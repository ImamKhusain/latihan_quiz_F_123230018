### 1. Penggabungan Matriks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Soal 1: Penggabungan Matriks A dan B secara Horizontal
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
hasil_gabungan = np.hstack((A, B))
print(f"Penggabungan Secara Horizontal: \n{hasil_gabungan}")
```

### 2. Penjumlahan Matriks

```python
# Soal 2: Operasi Penjumlahan Matriks A + B
hasil_penjumlahan = A + B
print(f"Hasil Penjumlahan A + B: \n{hasil_penjumlahan}")
```

### 3. Load data dari file CSV 'pengunjung_lab.csv' dan tampilkan datanya

```python
data = pd.read_csv("pengunjung_lab.csv")
print(data)
```

### 4. Data Preprocessing:  Cek tipe data dan jumlah baris serta kolom dalam dataset

```python
print(data.info())
```

### 5. Data Preprocessing: Ubah Column tahun menjadi String

```python
data['tahun'] = data['tahun'].astype('string')
print(data.info())
```

### 6. Data Preprocessing:  Buat semua data yang NaN pada data pengunjung, supaya diisi menjadi 1000.

```python
import numpy as np
data.replace(np.nan, 1000, inplace=True)
print(data)
```

### 7. Data Preprocessing:  Simpan data yang sudah diubah NaN nya ke file baru dengan nama pengunjung_cleaned.csv

```python
data.to_csv('pengunjung_cleaned.csv', index=False)
```

### 8. Visualisasi Data:  Tampilkan Grafik Garis Untuk membandingkan pengunjung dengan prediksi.

```python
# Load ulang data yang sudah dibersihkan
data_cleaned = pd.read_csv('pengunjung_cleaned.csv')

plt.figure(figsize=(10,5))
plt.plot(data_cleaned['tahun'], data_cleaned['pengunjung'], label='Pengunjung Rill', marker='o')
plt.plot(data_cleaned['tahun'], data_cleaned['prediksi'], label='Prediksi', marker='s')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pengunjung')
plt.title('Perbandingan Pengunjung Rill dan Prediksi')
plt.legend()
plt.grid()
plt.show()
```
